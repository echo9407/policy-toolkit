```python
import pandas as pd
```

```python
data = pd.read_csv("../data-results.csv")
data = data[data['results'] < 0.05]
```

```python
classes = []
# 1 = incentive
# 2 = dis
# 3 - neutral
for i in data['sentences'][len(classes):]:
    print(i)
    inp = input("What class?")
    if len(inp) > 0:
        classes.append(inp)
    else:
        inp = input("What class?")
        classes.append(inp)
```

```python
data['classes'] = classes
```

```python
data.to_csv("negative_data.csv")
```
